```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# YOLOv10 모델 불러오기 (COCO 80 classes 사전학습 weight)
model = YOLO("yolov10s.pt")   # yolov10m/l/x 도 가능

# 테스트용 이미지 다운로드 (터미널에서 wget 사용하거나, 브라우저에서 직접 다운로드)
# wget -O zidane.jpg https://github.com/ultralytics/yolov5/raw/master/data/images/zidane.jpg

# 객체 검출 (추론)
results = model("zidane.jpg", save=True, conf=0.25)

# 결과 이미지 확인 (Matplotlib)
result_img = results[0].plot()   # 탐지된 박스/라벨 그리기
plt.imshow(cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
